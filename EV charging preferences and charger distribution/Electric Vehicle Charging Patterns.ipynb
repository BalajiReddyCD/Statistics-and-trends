```python
def plot_pie_chart(data):
    plt.figure(figsize=(4,4))
    data['Charger Type'].value_counts().plot.pie(autopct='%1.1f%%') #plotting pie chart for charger type
    plt.title("Distribution of charger types")
    plt.ylabel('')
    plt.savefig("plot_pie_chart.png") #saving png figure
    plt.show()
    
plot_pie_chart(data)
```

```python
def plot_scatter(data):
    plt.figure(figsize=(8, 4))
    sns.scatterplot(data=data, x='Charging Duration (hours)', y='Energy Consumed (kWh)', 
                    hue='Vehicle Model', palette='plasma', s=100)
    
    plt.title('Energy Consumed vs Charging Duration by Vehicle Model')
    plt.xlabel('Charging Duration (hours)')        
    plt.ylabel('Energy Consumed (kWh)')     
    plt.legend(title='Vehicle Model')
    plt.savefig('scatter_plot.png')
    plt.show()

# Calling the function
plot_scatter(data) 
```

```python
def plot_heatmap(data):
    plt.figure(figsize=(8, 6))
    #Coorelation of matrix
    correlation_matrix = data[['Battery Capacity (kWh)', 'Charging Duration (hours)', 
                               'Charging Rate (kW)', 'Charging Cost (USD)', 
                               'State of Charge (Start %)', 'State of Charge (End %)', 
                               'Distance Driven (since last charge) (km)', 
                               'Temperature (°C)', 'Vehicle Age (years)']].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Heatmap of Key Features')
    plt.savefig('heat_map.png')
    plt.show()

plot_heatmap(data)
```
