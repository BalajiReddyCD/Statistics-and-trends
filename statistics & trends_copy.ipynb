```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#World Top Companies: Key Financial Analysis
yield_df = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Dividend_Yield.csv")
earnings_df = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Earnings.csv")
marketcap_df = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Market_Cap.csv")
peratio_df = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_P_E_ratio.csv")
revenue_df = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Revenue.csv")
```

```python
# Merging datasets on 'Name' column
# Merging datasets with suffixes to handle overlapping column names
merged_df1 = pd.merge(yield_df, earnings_df, on='Name', suffixes=('_yield', '_earnings'))
merged_df2 = pd.merge(merged_df1, marketcap_df, on='Name', suffixes=('', '_marketcap'))
merged_df3 = pd.merge(merged_df2, peratio_df, on='Name', suffixes=('', '_peratio'))
merged_df = pd.merge(merged_df3, revenue_df, on='Name', suffixes=('', '_revenue'))

print(merged_df.shape)
```

```python
#displaying the general information of the DataFrame
merged_df.info()
```

```python
#Summary of statistics of numerical column
merged_df.describe()
```

```python
#selecting top 10 companies by dividend_yield, Earnings, Revenue
top_10_companies = yield_df.nlargest(10, "dividend_yield_ttm")
top_10_companies1 = earnings_df.nlargest(10, "earnings_ttm")
top_10_companies2 = revenue_df.nlargest(10, "revenue_ttm")

#Defining figure with 2 Rows and 3 Columns
fig, axs = plt.subplots(2, 3, figsize=(15, 10))  # 2 rows, 3 columns

# Bar chart - Top 10 companies by dividend yield
axs[0, 0].bar(top_10_companies['Name'], top_10_companies['dividend_yield_ttm'], color="skyblue")
axs[0, 0].set_title('Top 10 companies by dividend yield')
axs[0, 0].tick_params(axis='x', rotation=45)

# Horizontal bar chart - Top 10 companies by earnings
axs[0, 1].barh(top_10_companies['Name'], top_10_companies1['earnings_ttm'], color="green")
axs[0, 1].set_title('Top 10 companies by Earnings')
axs[0, 1].set_xlabel('Earnings (TTM)')

# Scatter Plot - Dividend Yield vs Revenue
axs[0, 2].scatter(top_10_companies2['revenue_ttm'], top_10_companies['dividend_yield_ttm'], color="red")
axs[0, 2].set_title('Dividend Yield vs Revenue')
axs[0, 2].set_xlabel('Revenue (TTM)')
axs[0, 2].set_ylabel('Dividend Yield (%)')

# Line chart - Trend of earnings for top companies
axs[1, 0].plot(top_10_companies1['Name'], top_10_companies1['earnings_ttm'], marker='o', color="purple")
axs[1, 0].set_title('Trend of Earnings (TTM) for Top 10 Companies')
axs[1, 0].set_ylabel('Earnings (TTM)')
axs[1, 0].tick_params(axis='x', rotation=45)

# Pie chart - Market share by revenue
axs[1, 1].pie(top_10_companies2['revenue_ttm'], labels=top_10_companies2['Name'], autopct='%1.1f%%', colors=sns.color_palette("Set3"))
axs[1, 1].set_title('Market Share by Revenue (Top 10)')

# Heatmap - Correlation between key metrics
metrics_df = merged_df[['dividend_yield_ttm', 'earnings_ttm', 'revenue_ttm','marketcap','pe_ratio_ttm']].corr()
sns.heatmap(metrics_df, annot=True, cmap='coolwarm', ax=axs[1, 2])
axs[1, 2].set_title('Correlation Heatmap of Financial Metrics')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
