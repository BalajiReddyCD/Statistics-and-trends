```python
#importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Reading CSV files using pandas
yield_df     = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Dividend_Yield.csv")
earnings_df  = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Earnings.csv")
marketcap_df = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Market_Cap.csv")
peratio_df   = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_P_E_ratio.csv")
revenue_df   = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Revenue.csv")
```

```python
# Merging datasets on 'Name' column
# Merging datasets with suffixes to handle overlapping column names
merged_df1 = pd.merge(yield_df, earnings_df, on='Name', suffixes=('_yield', '_earnings'))
merged_df2 = pd.merge(merged_df1, marketcap_df, on='Name', suffixes=('', '_marketcap'))
merged_df3 = pd.merge(merged_df2, peratio_df, on='Name', suffixes=('', '_peratio'))
merged_df  = pd.merge(merged_df3, revenue_df, on='Name', suffixes=('', '_revenue'))
merged_df
```

```python
# Selecting top 10 companies by dividend yield, earnings, revenue
top_10_companies_yield = yield_df.nlargest(10, "dividend_yield_ttm")
top_10_companies_earnings = earnings_df.nlargest(10, "earnings_ttm")
top_10_companies_revenue = revenue_df.nlargest(10, "revenue_ttm")
top_10_companies_marketcap = marketcap_df.nlargest(10, "marketcap")
top_10_companies_pe_ratio = peratio_df.nlargest(10, "pe_ratio_ttm")
top_10_companies_revenue
```

# Visualization 1
### Bar Plots with Heatmap

```python
# Function to create a heatmap for the correlation between key metrics
def plot_correlation_heatmap(merged_df):
    plt.figure(figsize=(8, 6))  # Adjusting the figure size for the heatmap
    metrics_df = merged_df[['dividend_yield_ttm', 'earnings_ttm', 'revenue_ttm', 'marketcap', 'pe_ratio_ttm']].corr()
    sns.heatmap(metrics_df, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap of Financial Metrics')

    plt.savefig("bar_plot.png")  # Save the bar plot figure

# Plotting the heatmap
plot_correlation_heatmap(merged_df)
```

# Visualization 2
### Scatter Plots

```python
# Map countries to unique integer codes
top_10_companies_revenue['Country_Code'] = top_10_companies_revenue['country'].astype('category').cat.codes
```

```python
import matplotlib.pyplot as plt

# Function to create a scatter plot for Revenue vs Market Cap
def plot_revenue_vs_market_cap(top_10_companies_revenue, top_10_companies_marketcap):
    fig, ax = plt.subplots(figsize=(10, 6))  # figure size

    # Creating the scatter plot
    sc = ax.scatter(
        top_10_companies_revenue['revenue_ttm'],
        top_10_companies_marketcap['marketcap'],
        c=top_10_companies_revenue['Country_Code'],  # Color by country code
        cmap='plasma',  # Color map for distinguishing countries
        s=100  # Size of the points
    )

    # Set titles and labels
    ax.set_title('Revenue vs Market Cap', fontsize=16)
    ax.set_xlabel('Revenue (TTM)', fontsize=12)
    ax.set_ylabel('Market Cap ($B)', fontsize=12)

    # Adding a color bar to indicate the countries
    cbar = fig.colorbar(sc, ax=ax, orientation='vertical', fraction=0.05, pad=0.02)
    cbar.set_label('Country', fontsize=12)
    cbar.set_ticks(range(len(top_10_companies_revenue['Country_Code'].unique())))
    cbar.set_ticklabels(top_10_companies_revenue['Country_Code'].unique()) 

    # Showing grid lines
    ax.grid(True, linestyle='--', alpha=0.7)

    # Saving the plot
    plt.tight_layout()
    plt.savefig("revenue_vs_market_cap.png")
    plt.show()

# Calling the function to create the scatter plot
plot_revenue_vs_market_cap(top_10_companies_revenue, top_10_companies_marketcap)
```

### Histogram plots

```python
# Grouping by Country and calculating average's
average_yield = yield_df.groupby('Name')['dividend_yield_ttm'].mean().reset_index()
average_earnings = earnings_df.groupby('Name')['earnings_ttm'].mean().reset_index()
average_marketcap = marketcap_df.groupby('Name')['marketcap'].mean().reset_index()
average_revenue = revenue_df.groupby('Name')['revenue_ttm'].mean().reset_index()
```

# Visualization 4
### Pie Charts

```python
# Function to create a pie chart for revenue distribution
def plot_revenue_distribution(ax, revenues, labels):
    ax.pie(
        revenues,
        labels=labels,
        autopct='%1.1f%%',  # Format for percentage display
        colors=sns.color_palette("Set3")  # Color palette for the pie chart
    )
    ax.set_title('Revenue Distribution by Company')  # Title for the pie chart

# Main function to create the revenue distribution chart
def create_revenue_distribution_chart(top_10_companies_revenue):
    fig, ax = plt.subplots(figsize=(8, 8))  # Create a single subplot

    # Calling the plotting function for the revenue distribution
    plot_revenue_distribution(ax, top_10_companies_revenue['revenue_ttm'], top_10_companies_revenue['Name'])

    plt.tight_layout()  # Adjust layout
    plt.savefig("revenue_distribution_pie_chart.png")  # Save the pie chart figure
    plt.show()  # Show the pie chart

# Calling the function to create the revenue distribution chart
create_revenue_distribution_chart(top_10_companies_revenue)
```
