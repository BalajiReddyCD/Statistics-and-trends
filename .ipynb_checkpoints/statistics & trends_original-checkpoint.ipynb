# World Top Companies: Key Financial Analysis

```python
#importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Reading CSV files using pandas
yield_df     = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Dividend_Yield.csv")
earnings_df  = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Earnings.csv")
marketcap_df = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Market_Cap.csv")
peratio_df   = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_P_E_ratio.csv")
revenue_df   = pd.read_csv("C:\\Users\\BALA\\OneDrive - University of Hertfordshire\\Desktop\\Applied DataScience1\\Assignment 2 statistics and trends\\Companies_ranked_by_Revenue.csv")
```

```python
# Merging datasets on 'Name' column
# Merging datasets with suffixes to handle overlapping column names
merged_df1 = pd.merge(yield_df, earnings_df, on='Name', suffixes=('_yield', '_earnings'))
merged_df2 = pd.merge(merged_df1, marketcap_df, on='Name', suffixes=('', '_marketcap'))
merged_df3 = pd.merge(merged_df2, peratio_df, on='Name', suffixes=('', '_peratio'))
merged_df  = pd.merge(merged_df3, revenue_df, on='Name', suffixes=('', '_revenue'))
merged_df
```

```python
# Selecting top 10 companies by dividend yield, earnings, revenue
top_10_companies_yield = yield_df.nlargest(10, "dividend_yield_ttm")
top_10_companies_earnings = earnings_df.nlargest(10, "earnings_ttm")
top_10_companies_revenue = revenue_df.nlargest(10, "revenue_ttm")
top_10_companies_marketcap = marketcap_df.nlargest(10, "marketcap")
top_10_companies_pe_ratio = peratio_df.nlargest(10, "pe_ratio_ttm")
top_10_companies_revenue
```

# Visualization 1
### Bar Plots with Heatmap

```python
# Function to creating a top 10 companies by dividend yield
def plot_dividend_yield(top_10_companies_yield):
    plt.subplot(3, 3, 1)  # Setting the position for this subplot
    plt.bar(top_10_companies_yield['Name'], top_10_companies_yield['dividend_yield_ttm'], color="skyblue")
    plt.title('Top 10 Companies by Dividend Yield')  # Title for the plot
    plt.xticks(rotation=45)  # Rotating x-axis labels for better readability

# Function to creating a top 10 companies by earnings
def plot_earnings(top_10_companies_earnings):
    plt.subplot(3, 3, 2)  
    plt.bar(top_10_companies_earnings['Name'], top_10_companies_earnings['earnings_ttm'], color="green")
    plt.title('Top 10 Companies by Earnings')
    plt.xticks(rotation=45)

# Function to creating a top 10 companies by revenue
def plot_revenue(top_10_companies_revenue):
    plt.subplot(3, 3, 3)  
    plt.bar(top_10_companies_revenue['Name'], top_10_companies_revenue['revenue_ttm'], color="orange")
    plt.title('Top 10 Companies by Revenue')
    plt.xticks(rotation=45)

# Function to creating a top 10 companies by market capitalization
def plot_market_cap(top_10_companies_marketcap):
    plt.subplot(3, 3, 4) 
    plt.bar(top_10_companies_marketcap['Name'], top_10_companies_marketcap['marketcap'], color="purple")
    plt.title('Top 10 Companies by Market Cap')
    plt.xticks(rotation=45)

# Function to creating a top 10 companies by P/E ratio
def plot_pe_ratio(top_10_companies_pe_ratio):
    plt.subplot(3, 3, 5)  
    plt.bar(top_10_companies_pe_ratio['Name'], top_10_companies_pe_ratio['pe_ratio_ttm'], color="red")
    plt.title('Top 10 Companies by P/E Ratio')
    plt.xticks(rotation=45)

# Function to creating a heatmap for the correlation between key metrics
def plot_correlation_heatmap(merged_df):
    plt.subplot(3, 3, 6) 
    metrics_df = merged_df[['dividend_yield_ttm', 'earnings_ttm', 'revenue_ttm', 'marketcap', 'pe_ratio_ttm']].corr()
    sns.heatmap(metrics_df, annot=True, cmap='coolwarm', cbar_kws={"shrink": .8}, ax=plt.gca())
    plt.title('Correlation Heatmap of Financial Metrics')

# Main function to setting up the figure and calling each plotting function
def create_visualizations(top_10_companies_yield, top_10_companies_earnings, top_10_companies_revenue, 
                          top_10_companies_marketcap, top_10_companies_pe_ratio, merged_df):
    plt.figure(figsize=(25, 15))  # Assigning the figure size

    # Calling each plotting function to create the respective plots
    plot_dividend_yield(top_10_companies_yield)
    plot_earnings(top_10_companies_earnings)
    plot_revenue(top_10_companies_revenue)
    plot_market_cap(top_10_companies_marketcap)
    plot_pe_ratio(top_10_companies_pe_ratio)
    plot_correlation_heatmap(merged_df)

    plt.tight_layout()  # Adjusting spacing to prevent overlap
    plt.show()  # Showing all the plots together

# calling to create all the visualizations
create_visualizations(top_10_companies_yield, top_10_companies_earnings, top_10_companies_revenue, 
                      top_10_companies_marketcap, top_10_companies_pe_ratio, merged_df)
```

# Visualization 2
### Scatter Plots

```python
# Map countries to unique integer codes
top_10_companies_revenue['Country_Code'] = top_10_companies_revenue['country'].astype('category').cat.codes
```

```python
# Function to plot Revenue vs P/E Ratio
def plot_revenue_vs_pe(ax):
    # Creating the scatter plot using revenue and P/E ratio, with colors by country
    sc = ax.scatter(
        top_10_companies_revenue['revenue_ttm'],
        top_10_companies_pe_ratio['pe_ratio_ttm'],
        c=top_10_companies_revenue['Country_Code'],
        cmap='plasma',  # Choosing a color map that makes it easy to distinguish countries
        s=30  # Setting a good size for the points so theyâ€™re easy to see
    )
    ax.set_title('Revenue vs P/E Ratio')  # Set title for this plot
    ax.set_xlabel('Revenue (TTM)')        # Label for the x-axis
    ax.set_ylabel('P/E Ratio')            # Label for the y-axis
    return sc  # Returning the scatter plot for the color bar later

# Function to plot Revenue vs Earnings
def plot_revenue_vs_earnings(ax):
    ax.scatter(
        top_10_companies_revenue['revenue_ttm'],
        top_10_companies_earnings['earnings_ttm'],
        c=top_10_companies_revenue['Country_Code'],
        cmap='plasma',
        s=30
    )
    ax.set_title('Revenue vs Earnings')
    ax.set_xlabel('Revenue (TTM)')
    ax.set_ylabel('Earnings (TTM)')

# Function to plot Revenue vs Market Cap
def plot_revenue_vs_marketcap(ax):
    ax.scatter(
        top_10_companies_revenue['revenue_ttm'],
        top_10_companies_marketcap['marketcap'],
        c=top_10_companies_revenue['Country_Code'],
        cmap='plasma',
        s=30
    )
    ax.set_title('Revenue vs Market Cap')
    ax.set_xlabel('Revenue (TTM)')
    ax.set_ylabel('Market Cap')

# Function to plot Earnings vs Dividend Yield
def plot_earnings_vs_dividend_yield(ax):
    ax.scatter(
        top_10_companies_earnings['earnings_ttm'],
        top_10_companies_yield['dividend_yield_ttm'],
        c=top_10_companies_revenue['Country_Code'],
        cmap='plasma',
        s=30
    )
    ax.set_title('Earnings vs Dividend Yield')
    ax.set_xlabel('Earnings (TTM)')
    ax.set_ylabel('Dividend Yield (%)')

# Main function to create the whole figure with subplots
def create_plots():
    # Setting up a grid of 2x2 plots, making sure thereâ€™s enough space for each
    fig, axs = plt.subplots(2, 2, figsize=(15, 10))

    # Calling each plot function, passing in the right subplot position
    sc1 = plot_revenue_vs_pe(axs[0, 0])    # Top-left
    plot_revenue_vs_earnings(axs[0, 1])    # Top-right
    plot_revenue_vs_marketcap(axs[1, 0])   # Bottom-left
    plot_earnings_vs_dividend_yield(axs[1, 1])  # Bottom-right

    # Adding a single color bar for all plots, labeling countries
    cbar = fig.colorbar(sc1, ax=axs, orientation='vertical', fraction=0.05, pad=0.02)
    cbar.set_label('Country')  # Set label for the color bar
    cbar.set_ticks(range(len(top_10_companies_revenue['country'].unique())))
    cbar.set_ticklabels(top_10_companies_revenue['country'].unique())  # Match each color to a country

    # Adjusting the spacing so titles and labels donâ€™t overlap, leaving space for the color bar
    plt.subplots_adjust(left=0.1, right=0.85, top=0.9, bottom=0.1, wspace=0.4, hspace=0.4)
    plt.show()  # Display all plots at once

# Running the main function to generate the visualizations
create_plots()
```

# Visualization 3
### Horizontal Bar plots

```python
# Grouping by Country and calculating average's
average_yield = yield_df.groupby('Name')['dividend_yield_ttm'].mean().reset_index()
average_earnings = earnings_df.groupby('Name')['earnings_ttm'].mean().reset_index()
average_marketcap = marketcap_df.groupby('Name')['marketcap'].mean().reset_index()
average_revenue = revenue_df.groupby('Name')['revenue_ttm'].mean().reset_index()
```

```python
# Sorting by average in descending order and selecting the top 10
top_10_countries1 = average_yield.sort_values(by='dividend_yield_ttm', ascending=False).head(10)
top_10_countries2 = average_earnings.sort_values(by='earnings_ttm', ascending=False).head(10)
top_10_countries3 = average_marketcap.sort_values(by='marketcap', ascending=False).head(10)
top_10_countries4 = average_revenue.sort_values(by='revenue_ttm', ascending=False).head(10)
```

```python
# Preparing the data for plotting
countries = top_10_countries1['Name']

average_yields  = top_10_countries1['dividend_yield_ttm']
average_earning = top_10_countries2['earnings_ttm']
average_market  = top_10_countries3['marketcap']
average_revenue = top_10_countries4['revenue_ttm']
```

```python
# Function to create a average dividend yield by country
def plot_average_dividend_yield(ax, countries, average_yields):
    ax.barh(countries, average_yields, color='skyblue')  # Creating a horizontal bar chart
    ax.set_title('Top 10 Countries by Average Dividend Yield')  # Title for the subplot
    ax.set_xlabel('Average Dividend Yield (%)')  # X-axis label
    ax.grid(axis='x')  # Adding grid lines along the x-axis for better readability

# Function to create a average earnings by country
def plot_average_earnings(ax, countries, average_earning):
    ax.barh(countries, average_earning, color='green') 
    ax.set_title('Top 10 Countries by Average Earnings')  
    ax.set_xlabel('Average Earnings')  
    ax.grid(axis='x') 

# Function to create a average market cap by country
def plot_average_market_cap(ax, countries, average_market):
    ax.barh(countries, average_market, color="lightcoral") 
    ax.set_title('Top 10 Countries by Average Market Cap')  
    ax.set_xlabel('Average Market Cap ($B)')  
    ax.grid(axis='x')  

# Function to create a average revenue by country
def plot_average_revenue(ax, countries, average_revenue):
    ax.barh(countries, average_revenue, color="salmon")  
    ax.set_title('Top 10 Countries by Average Revenue')  
    ax.set_xlabel('Average Revenue ($B)')  
    ax.grid(axis='x')  

# Main function 
def create_country_charts(countries, average_yields, average_earning, average_market, average_revenue):
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Creating a 2*2 grid of subplots

    # Call each plotting function for the respective subplot
    plot_average_dividend_yield(axs[0, 0], countries, average_yields)
    plot_average_earnings(axs[0, 1], countries, average_earning)
    plot_average_market_cap(axs[1, 0], countries, average_market)
    plot_average_revenue(axs[1, 1], countries, average_revenue)

    # Adjusting the layout to prevent overlap
    plt.tight_layout()
    plt.show() 

# calling to create all the country charts
create_country_charts(countries, average_yields, average_earning, average_market, average_revenue)
```

# Visualization 4
### Pie Charts

```python
# Function to create a pie chart for revenue distribution
def plot_revenue_distribution(ax, revenues, labels):
    ax.pie(
        revenues,
        labels=labels,
        autopct='%1.1f%%',  # Format for percentage display
        colors=sns.color_palette("Set3")  # Color palette for the pie chart
    )
    ax.set_title('Revenue Distribution by Company')  # Title for the pie chart

# Function to create a for market capitalization distribution
def plot_market_cap_distribution(ax, market_caps, labels):
    ax.pie(
        market_caps,
        labels=labels,
        autopct='%1.1f%%', 
        colors=sns.color_palette("Set2")  
    )
    ax.set_title('Market Cap Distribution by Company')  

# Main function 
def create_distribution_charts(top_10_companies_revenue, top_10_companies_marketcap):
    fig, axs = plt.subplots(2, 1, figsize=(10, 10))  # Create a 2*1 grid of subplots

    # Calling each plotting function for the respective subplot
    plot_revenue_distribution(axs[0], top_10_companies_revenue['revenue_ttm'], top_10_companies_revenue['Name'])
    plot_market_cap_distribution(axs[1], top_10_companies_marketcap['marketcap'], top_10_companies_marketcap['Name'])

    plt.tight_layout() 
    plt.show()  

# calling to create both distribution charts
create_distribution_charts(top_10_companies_revenue, top_10_companies_marketcap)
```
